```python
import sys, math, os
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt    
import numpy as np
import random
import math
from scipy.stats import logistic

%matplotlib inline
```

# Introduction

This notebook serves to answer the "bonus" question:

> Generate 10,000 random numbers (i.e. sample) from a Logistic distribution with parameters “location" = 10 and “scale” = 2.

```python
# GENERATE SAMPLE
n, loc, scale = 10000, 10, 2

sample = []
for i in xrange(n):
    rand = random.random()
    sample.append(loc + scale * math.log(rand / (1-rand)))
```

```python
# CHECK SAMPLE AGAINST PDF
plt.figure(figsize=(17,4))
# fig, ax = plt.subplots(1, 1)

x = np.linspace(logistic.ppf(0.01, loc=loc, scale=scale),
                logistic.ppf(0.99, loc=loc, scale=scale), n)
plt.plot(x, logistic.pdf(x, loc=loc, scale=scale), lw=3, alpha=0.8, label='logistic pdf')


plt.hist(sample, normed=True, alpha=0.5, label='sample',bins=100)
plt.legend(loc='best', frameon=False)
plt.savefig('../reports/figures/6.0_bonus-0.svg')
```
